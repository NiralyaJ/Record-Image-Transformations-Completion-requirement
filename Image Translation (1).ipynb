# **Image Translation**

Write a Python program using OpenCV to perform image translation on the **"Given image"**.

The program should:

1. Import `cv2`, `numpy`, and `matplotlib.pyplot`.
2. Read the image **"given** in color mode.
3. Create a **translation matrix** that moves the image  
   **50 pixels to the right** and **80 pixels down**.
4. Apply the translation
5. Display the following using a **1 × 2 layout** in Matplotlib:  
   - **Original Image**  
   - **Translated Image**


```python
# Import 

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image using OpenCV
image = cv2.imread('baseball.jpg')

```

```python

image.shape
```

```python
# Display the image using Matplotlib

# Display the images.
plt.imshow(image[:,:,::-1])
plt.title('Original Image')
plt.show()
```

```python
# i) Image Translation
tx, ty = 100, 200  # Translation factors (shift by 100 pixels horizontally and 50 vertically)
M_translation = np.float32([[1, 0, tx], [0, 1, ty]])  # Translation matrix: 
# [1, 0, tx] - Horizontal shift by tx
# [0, 1, ty] - Vertical shift by ty
translated_image = cv2.warpAffine(image, M_translation, (636, 438)) 
```

```python
plt.imshow(translated_image[:,:,::-1])
plt.title("Translated Image")
plt.axis('on')
plt.show()
```

The program should:

1. Scale the image to:
   - **0.5 times** its original size (downscaling)
   - **2 times** its original size (upscaling)

# **Image Shearing**

Write a Python program using OpenCV to perform image shearing on the **given image**.

The program should:

1. Create a **shearing transformation matrix** to apply:
   - **Horizontal shearing**  
   - **Vertical shearing**
2. Apply the shearing transformations.
3. Display the following using a **1 × 3 layout** in Matplotlib:  
   - **Original Image**  
   - **Horizontally Sheared Image**  
   - **Vertically Sheared Image**


```python
# Image Shearing
shear_matrix = np.float32([[1, 0.5, 0], [0.5, 1, 0]])  
sheared_image = cv2.warpAffine(image, shear_matrix, (636, 438))
```

# **Image Reflection**

Write a Python program using OpenCV to perform image reflection on the **given image**.

The program should:

1. Generate the following reflected versions of the image:
   - **Horizontal Reflection** (flip left ↔ right)  
   - **Vertical Reflection** (flip top ↔ bottom)  
   - **Both-axis Reflection**  
2. Use `cv2.flip` with appropriate flip codes.
3. Display the following using a **2 × 2 layout** in Matplotlib:  
   - **Original Image**  
   - **Horizontally Reflected Image**  
   - **Vertically Reflected Image**

```python
# Image Reflection
reflected_image = cv2.flip(image, 2)  # Flip the image horizontally (1 means horizontal flip)

# flip: 1 means horizontal flip, 0 would be vertical flip, -1 would flip both axes
```

```python
# Show original image 
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image[:, :, ::-1])
plt.title("Original Image")
plt.axis('off')

# Show reflected image 
plt.subplot(1, 2, 2)
plt.imshow(reflected_image[:,:,::-1])
plt.title("Reflected Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **Image Rotation**

Write a Python program using OpenCV to perform **image rotation** on the **given image**.

The program should:

1. Rotate the image by:  
   - **45 degrees**  
   - **90 degrees**
2. Use `cv2.getRotationMatrix2D` and `cv2.warpAffine`.
3. Display the following using a **1 × 3 layout** in Matplotlib:  
   - **Original Image**  
   - **45° Rotated Image**  
   - **90° Rotated Image**


```python
# Image Rotation
(height, width) = image.shape[:2]  # Get the image height and width
angle = 45  # Rotation angle in degrees (rotate by 45 degrees)
center = (width // 2, height // 2)  # Set the center of rotation to the image center
M_rotation = cv2.getRotationMatrix2D(center, angle, 1)  # Get the rotation matrix
# getRotationMatrix2D: Takes the center of rotation, angle, and scale factor (1 means no scaling)
rotated_image = cv2.warpAffine(image, M_rotation, (width, height))  # Apply rotation
```

```python
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))  # Display the rotated image
plt.title("Rotated Image")  # Set title
plt.axis('off')
```
